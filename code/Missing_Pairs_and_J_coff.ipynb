```python
#Total_London_Checkin = pd.read_csv('/home/2019/nyu/fall/das968/citadel_comp/Total_London_Checkin.csv',index_col=0)
Total_London_Checkin = pd.read_csv(r'C:\Users\Darshan\Desktop\citadel_datathon\Total_London_Checkin.csv',index_col=0) 
```

```python
venue_location_df = Total_London_Checkin[['Venue ID', 'Latitude', 'Longitude', 'Category']].drop_duplicates().reset_index(drop=True)
venue_location_df.loc[venue_location_df['Category']=="Caf\x1a\x1a", "Category"] = "Cafe"
venue_location_df.loc[venue_location_df['Category']=="Caf\x1a", "Category"] = "Cafe"
#venue_location_df['MainCategory'] = venue_location_df['Category'].apply(getMainCategory)
```

```python
London_Olympic_Checkin = pd.read_csv(r'C:\Users\Darshan\Desktop\citadel_datathon\London_Olympics_Checkin.csv',index_col=0)  
venue_id_list = London_Olympic_Checkin.groupby(['Venue ID']).count()['User ID'].index[(London_Olympic_Checkin.groupby(['Venue ID']).count()['User ID'].values >= 3)]
```

```python
mask = venue_location_df.loc[venue_location_df['Venue ID'].isin(venue_id_list)]['Category'].value_counts().values > 1
subCat_list = venue_location_df.loc[venue_location_df['Venue ID'].isin(venue_id_list)]['Category'].value_counts().index[mask]
```

```python
tp_list = subCat_list
tv_list = subCat_list
```

```python
pairs_tv_tp = []
for i in range(len(tp_list)-1):
    for j in range(i+1,len(tv_list)):
        pairs_tv_tp.append((tp_list[i],tv_list[j]))
```

```python
len(pairs_tv_tp)
```

```python
j_table_231 = pd.read_csv(r'C:\Users\Darshan\Desktop\citadel_datathon\pairs_jcoff_231.txt',delimiter="#",header=None)  
j_table_1750 = pd.read_csv(r'C:\Users\Darshan\Desktop\citadel_datathon\pairs_jcoff_1750.txt',delimiter="#",header=None) 
j_table_200 = pd.read_csv(r'C:\Users\Darshan\Desktop\citadel_datathon\pairs_jcoff_200.txt',delimiter="#",header=None)  
j_table = pd.concat([j_table_231,j_table_1750,j_table_200])
j_table.columns = ["j_coef", 'tp', 'tv']

df_jr = pd.DataFrame()
df_jr ["j_coef"] = j_table['j_coef']
df_jr['tp'] = j_table['tv']
df_jr['tv'] = j_table['tp']

j_table_final = pd.concat([j_table,df_jr])
```

```python
pairs_from_IBM = []

for i in range(len(j_table_final)):
    if(j_table_final['tp'].iloc[i] != j_table_final['tv'].iloc[i]):
        pairs_from_IBM.append((j_table_final['tp'].iloc[i],j_table_final['tv'].iloc[i]))
```

```python
pairs_present = set()   
for p in pairs_from_IBM:    
    if (p in pairs_tv_tp):
        pairs_present.add(p)
```

```python
missing_pairs = []
for m in pairs_tv_tp:
    if not (m in pairs_present):
        missing_pairs.append(m)
```
